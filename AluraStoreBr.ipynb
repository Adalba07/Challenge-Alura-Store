### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
# Lista com os valores de faturamento
precos = [219.08, 256.35, 279.51, 1009.99, 446.99]

# Soma total do faturamento
faturamento_total = sum(precos)

# Exibir o resultado
print(f"Faturamento total: R${faturamento_total:.2f}")
```

# 2. Vendas por Categoria


```python
# prompt: calcular a quantidade de produtos vendidos por categoria em cada loja. A ideia é agrupar os dados por categoria e contar o número de vendas de cada tipo, mostrando as categoria mais populares de cada loja

def analisar_vendas_por_categoria(df, nome_loja):
    vendas_por_categoria = df.groupby('categoria')['id'].count().sort_values(ascending=False)
    print(f'\nVendas por categoria na {nome_loja}:')
    print(vendas_por_categoria)
    return vendas_por_categoria

# Analisar as vendas por categoria de cada loja
vendas_loja1 = analisar_vendas_por_categoria(loja, 'Loja 1')
vendas_loja2 = analisar_vendas_por_categoria(loja2, 'Loja 2')
vendas_loja3 = analisar_vendas_por_categoria(loja3, 'Loja 3')
vendas_loja4 = analisar_vendas_por_categoria(loja4, 'Loja 4')
```

# 3. Média de Avaliação das Lojas

```python
# prompt:  vamos calcular a média das avaliações dos clientes para cada loja. O objetivo é entender a satisfação dos clientes com os produtos vendidos.

def calcular_media_avaliacoes(df, nome_loja):
    """Calcula a média das avaliações dos clientes para uma loja."""
    media_avaliacoes = df['avaliacao'].mean()
    print(f'\nMédia de avaliações da {nome_loja}: {media_avaliacoes:.2f}')
    return media_avaliacoes

# Calcular a média de avaliações de cada loja
media_loja1 = calcular_media_avaliacoes(loja, 'Loja 1')
media_loja2 = calcular_media_avaliacoes(loja2, 'Loja 2')
media_loja3 = calcular_media_avaliacoes(loja3, 'Loja 3')
media_loja4 = calcular_media_avaliacoes(loja4, 'Loja 4')
```

# 4. Produtos Mais e Menos Vendidos

```python
def analisar_produtos_vendidos(df, nome_loja):
    """Identifica os produtos mais e menos vendidos em uma loja."""
    # Check if 'produto' column exists, if not, try 'Produto'
    column_name = 'produto' if 'produto' in df.columns else 'Produto'
    vendas_por_produto = df.groupby(column_name)['id'].count().sort_values(ascending=False)
    mais_vendido = vendas_por_produto.index[0]
    menos_vendido = vendas_por_produto.index[-1]
    quantidade_mais_vendido = vendas_por_produto.iloc[0]
    quantidade_menos_vendido = vendas_por_produto.iloc[-1]

    print(f"\nProdutos mais e menos vendidos na {nome_loja}:")
    print(f"Mais vendido: {mais_vendido} ({quantidade_mais_vendido} unidades)")
    print(f"Menos vendido: {menos_vendido} ({quantidade_menos_vendido} unidades)")

    return mais_vendido, menos_vendido, quantidade_mais_vendido, quantidade_menos_vendido
```
